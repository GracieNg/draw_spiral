```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# generate some sample data
# _ used to ignore the output label as want to try unsupervised
# by default this returns 2 features, if you need more then use eg n_features=n where n=1,2,3,4....n
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.05, random_state=42)

# plot the data before clustering
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], color='gray', marker='o', edgecolors='k')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Before Clustering")
plt.grid(True)

# K-Means Clustering with 3 clusters; k is number of clusters
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X)

# Cluster assignment
labels = kmeans.labels_
centroids = kmeans.cluster_centers_ #mean of the points of that cluster

# Visualize after cluster
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', edgecolors='k')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label="Centroids")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("After K-Means Clustering")
plt.legend()
plt.grid(True)

# Show both plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# generate some sample data
# _ used to ignore the output label as want to try unsupervised
# by default this returns 2 features, if you need more then use eg n_features=n where n=1,2,3,4....n
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.05, random_state=42)

# visualize clustering for different K values
def plot_kmeans(k):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    labels = kmeans.labels_
    centroids = kmeans.cluster_centers_

    # plot clustering outputs
    plt.figure(figsize=(6, 5))
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', edgecolors='k')
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label="Centroids")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.title(f"K-Means Clustering (K={k})")
    plt.legend()
    plt.grid(True)
    plt.show()

# try different k values
for k in [2, 3, 4, 5]:  # Try different values of K
    plot_kmeans(k)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# generate some sample data
# _ used to ignore the output label as want to try unsupervised
# by default this returns 2 features, if you need more then use eg n_features=n where n=1,2,3,4....n;
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=1.05, random_state=42)

# plot the data before clustering
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], color='gray', marker='o', edgecolors='k')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Before Clustering")
plt.grid(True)

# compute WCSS (Within-Cluster Sum of Squares) for different K values
wcss = []
# checking k from 1 to 10
K_values = range(1, 11)

for k in K_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_) # inertia = WCSS

# plot WCSS vs. K (Elbow Curve: optimal k is found at the "elbow", where the WCSS stops decreasing rapidly)
plt.figure(figsize=(8,5))
plt.plot(K_values, wcss, marker='o', linestyle='--', color='b')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS (Within-Cluster Sum of Squares)")
plt.title("Elbow Method for Optimal K")
plt.grid(True)

plt.xticks(K_values)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score, silhouette_score, adjusted_rand_score
from scipy.stats import mode
from sklearn.decomposition import PCA

# load Iris dataset; 150 flower samples with 4 features (sepal length, sepal width, petal length, petal width).
iris = load_iris()
# X features; there are four features
X = iris.data
# y_true is output labels (0, 1, 2 for 3 species);
# We are not going to use this for the K-Means but for evaluating as how well did the model do.
# k means should be best with k=3; let us see
y_true = iris.target

# let us have a look at data before clustring
plt.figure(figsize=(7, 5))
plt.scatter(X[:, 0], X[:, 1], c=y_true, cmap='viridis', marker='o', edgecolors='k')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Iris Dataset Before Clustering (Actual Labels)")
plt.grid(True)
plt.show()

print(X)
print(y_true)

```

```python
# k-means
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X)
y_kmeans = kmeans.labels_
print(y_kmeans)

```

```python
# mapping cluster labels to actual labels
def map_clusters_to_labels(y_kmeans, y_true):
    mapped_labels = np.zeros_like(y_kmeans)
    print(mapped_labels)
    for cluster in np.unique(y_kmeans):
        mask = (y_kmeans == cluster)
        print(mask)
        print(y_true[mask])
        mapped_labels[mask] = mode(y_true[mask])[0]
        print(mode(y_true[mask])[0])
    return mapped_labels

y_kmeans_mapped = map_clusters_to_labels(y_kmeans, y_true)
print(y_kmeans_mapped)

```

```python
# compute metrics
accuracy = accuracy_score(y_true, y_kmeans_mapped)
silhouette = silhouette_score(X, y_kmeans)
ari = adjusted_rand_score(y_true, y_kmeans)

print(f"Adjusted Accuracy: {accuracy:.2f}")  # adjusted for best match
print(f"Silhouette Score: {silhouette:.2f}")  # measures how well-separated the clusters are (clustering separation quality); lower score:overlap
print(f"Adjusted Rand Index: {ari:.2f}")  # checks how similar K-Means clusters are to the actual classes (similarity to real classes); lower score: misclassification

# visualize using pca (dimension reduction; 4 dimensional data to 2d for ease of visualization )
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(7,5))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans, cmap='viridis', marker='o', edgecolors='k')
plt.scatter(pca.transform(kmeans.cluster_centers_)[:, 0],
            pca.transform(kmeans.cluster_centers_)[:, 1],
            c='red', marker='X', s=200, label="Centroids")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.title("K-Means Clustering on Iris Dataset")
plt.legend()
plt.grid(True)
plt.show()
```
